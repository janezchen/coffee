# ECONOMIC FACTORS

In the global coffe market, the major importers of coffee are developed countries such as Japan, USA and many European countries like Germany, France, etc. On the other hand, the major producers and exporters of coffee are developing nations like Brazil, Vietnam and Indonesia. Furthermore, the coffee farmers in developing countries are low income earners dependent on coffee prices for making a living. Given this information it would make sense that the economic well being of both exporting and importing countries would play an important role in determining the demand and supply of coffee and ultimately its price in the global market.

In order to investigate this idea, we will consider economic factors such as the GDP per capita and the income share of the richest/poorest 20% of importing and exporting countries.

### Gather Data

First we must gather relevant data for our investigation. The world bank has gdp per capita data as well as income share data for many countries going back several decades. We will only consider data for the major importers and exporters for this analysis because they are the most important drivers of observable changes in the global coffee market.

```python
# get coffee production and consumption info
productionDf = pd.read_csv("production.csv")
consumptionDf = pd.read_csv("global_consumption.csv")
prod_countries, cons_countries = productionDf.Country, consumptionDf.Country

# calculate mean production and consumption for each country
prod, cons = productionDf[[str(yr) for yr in range(1990,2013)]], consumptionDf[[str(yr) for yr in range(1990,2013)]]
mean_prod = prod.apply(lambda lst: map(lambda elt: int(elt.replace(',','')),lst)).mean(axis=1)
mean_cons = cons.apply(lambda lst: map(lambda elt: int(str(elt).replace(',','').replace('nan','0')),lst)).mean(axis=1)

# countries to be considered
major_importers = list(cons_countries[mean_cons > 4500])
major_importers.remove('Total')
major_exporters = list(prod_countries[mean_prod > 4500])

print "Biggest coffee importers",major_importers
print "Biggest coffee exporters",major_exporters
```

Now we acquire the income shares data.

```python
# Income share of the richest/poorest 20% of exporting and importing countries
poorestdf = pd.read_csv("income_Share/countries_incomeshare_poorest20percent.csv")[['Country Name']+[str(yr) for yr in range(1990,2013)]]
richestdf = pd.read_csv("income_Share/countries_incomeshare_richest20percent.csv")[['Country Name']+[str(yr) for yr in range(1990,2013)]]

richExp, richImp = richestdf.set_index('Country Name').loc[major_exporters], richestdf.set_index('Country Name').loc[major_importers]
poorExp, poorImp = poorestdf.set_index('Country Name').loc[major_exporters], poorestdf.set_index('Country Name').loc[major_importers]

poorExp
```

There seems to be quite a bit of missing information in our data. Let's fill these by using the average income between 1990-2012. We write a function to fill up these missing data.

```python
def clean(df):
    cleanrow = lambda lst: map(lambda elt: np.nanmean(lst) if np.isnan(elt) else elt, lst)
    return df.apply(cleanrow, axis=1)
poorExp, richExp = clean(poorExp), clean(richExp)
poorImp, richImp = clean(poorImp), clean(richImp)
poorExp
```

Now let's gather and clean GDP per capita data

```python
# get GDP per capita data
gdpdf = pd.read_csv("gdp_percapita.csv")[['Country Name']+[str(yr) for yr in range(1990,2013)]]
gdpExp, gdpImp = clean(gdpdf.set_index('Country Name').loc[major_exporters]), clean(gdpdf.set_index('Country Name').loc[major_importers])
print "GDP PER CAPITA FOR COFFEE IMPORTING COUNTRIES"
gdpImp
```

```python
def filter_rows(df, start, end):
    return df.loc[filter(lambda i: int(i.year) >= start and int(i.year) <= end, df.index)]

pricedf = pd.DataFrame.from_csv('yearly_prices.csv')
pricedf.index = pd.to_datetime(pricedf.index)

prices = filter_rows(pricedf, 1990, 2012)['ICO composite indicator']

avg_gdpImp = gdpImp.mean(axis=0)
avg_gdpImp.name = 'Average GDP per capita of importing Countriess'
sns.regplot(prices, avg_gdpImp)
print "Correlation between prices and average GDP of coffee importing countries",avg_gdpImp.corr(prices)
```

```python
avg_gdpExp = gdpExp.mean(axis=0)
avg_gdpExp.name = 'Average GDP per capita of exporting Countriess'
sns.regplot(prices, avg_gdpExp)
print "Correlation between prices and average GDP of coffee importing countries", avg_gdpExp.corr(prices)
```

There seems to be a strong correlation between coffee prices and the GDP per capita of importing countries and an even stronger correlation with exporting countries. This trend is understandable because if the economies of importing countries are doing well, we can expect demand for coffee to rise leading to an increase in coffee prices, ultimately resulting in increased revenue for the exporting countries. 

To investigate this trend in more detail let's take a look at how coffee prices are affected by the rich in importing countries, who are big drivers of demand, and the poor in the exporting countries, who mostly constitute the farmers. 

```python
# average income share of richest 20% in importing countries
avg_richImp = richImp.mean(axis=0)
avg_richImp.name = 'Average income share of richest 20% in importing Countriess'
sns.regplot(prices, avg_richImp)
print "Correlation between prices and average average income share of richest 20% in coffee importing countries:" 
print avg_richImp.corr(prices)
```

```python
# average income share of richest 20% in importing countries
avg_poorExp = poorExp.mean(axis=0)
avg_poorExp.name = 'Average income share of poorest 20% in exporting Countriess'
sns.regplot(prices, avg_poorExp)
print "Correlation between prices and average average income share of poorest 20% in coffee exporting countries:" 
print avg_poorExp.corr(prices) 
```

As we would expect an increase in the income share of the rich should account for a rise in coffee demand and hence an increase in prices, however to our suprise this does not constitute an increase in income share for the poor farmers producing coffee in exporting nations. Correlations do not tell us anything about causal relationships, therefore at this point in time all we can do provide hypotheses for this observation. One hypothesis is that the price increases only serve to benefit the rich in these exporting countries and this would explain the reduction in income share for the poor that is implied by our above analysis. It may also be the case that all the missing information in our income share data is causing our unusual results.

To investigate our hypothesis let us observe the relationship betweeen income share of the rich in the exporting countries and the price of coffee. To test whether our data might be the problem we'll also look at the relationship between income shares and average consumption in importing countries.

```python
# average income share of richest 20% in exporting countries
avg_richExp = richExp.mean(axis=0)
avg_richExp.name = 'Average income share of richest 20% in importing Countriess'
sns.regplot(prices, avg_richExp)
print "Correlation between prices and average average income share of richest 20% in coffee importing countries:" 
print avg_richExp.corr(prices)
```

This tells us that the income share of the rich also goes down, which seems to imply the the middle 60% of the exporting countries benefit the most from a rise in coffee prices. To see whether our income share data is meaningful, let us now look at it's relationship with consumption of importing countries.

```python
# get coffee consumption of major importers 
consImp = consumptionDf[['Country']+[str(yr) for yr in range(1990,2013)]].set_index('Country').loc[major_importers]
consImp = consImp.apply(lambda lst: map(lambda elt: float(str(elt).replace(',','')) if str(elt) != 'nan' else np.nan,lst))
consImp = clean(consImp)
consImp.head()
```

```python
# calculate average consumption of importing countries
avg_cons = consImp.mean(axis=0)
avg_richImp = richImp.mean(axis=0)

avg_cons.name = 'Average coffee consumption of importing Countriess'
avg_richImp.name = 'Average income share of richest 20% in importing Countriess'

sns.regplot(avg_cons, avg_richImp)
print "Correlation between rich income and average consumption of coffee importing countries",avg_cons.corr(avg_richImp)
```

There is a positive correlation but it is not quite as much as we would expect. 

We can conclude from all the analysis we have performed above, that the GDP of exporting countries would be a great feature with which to build a predictive model. However due to the lack of information in the income share data we collected, we can be less confident about a predictive model using this as a feature.
